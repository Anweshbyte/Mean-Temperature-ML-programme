```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

col_list = ["Year", "Mean"]
data = pd.read_csv('temp.csv', usecols = col_list)
print(data.shape)
```

```python
 x = data['Year'].values
 y = data['Mean'].values

 plt.scatter(x, y, c = '#ef5423', label = 'Scatter Plot')
 plt.legend()
 plt.show()
```

```python
meanx = np.mean(x)
meany = np.mean(y)

m = len(x)

n = 0
d = 0

for i in range(m):
    n += (x[i] - meanx )*(y[i] - meany)
    d += (x[i] - meanx)**2

b1 = n/d
b0 = meany - (b1*meanx)
print(b1, b0)
```

```python
maxX = np.max(x) 
minX = np.min(x) 

X = np.linspace(minx, maxx, 40)
Y = b0 + b1*X
plt.plot(X , Y, color = '#58b970', label = 'Regression line')
plt.scatter(x, y, c = '#ef5423', label = 'Scatter Plot')

plt.xlabel('YEAR')
plt.ylabel('MEAN')
plt.legend()
plt.show()
```
